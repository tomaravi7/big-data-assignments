```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
```

```python
data=pd.read_csv('https://raw.githubusercontent.com/AnjulaMehto/MCA/main/company_sales_data.csv')

print(data)
plt.plot(data['month_number'],data['total_profit'])
plt.xlabel('Month')
plt.ylabel('Total Profit')
plt.show()
```

```python
plt.plot(data['month_number'],data['facecream'],label='facecream')
plt.plot(data['month_number'],data['facewash'],label='facwash')
plt.plot(data['month_number'],data['toothpaste'],label='toothpaste')
plt.plot(data['month_number'],data['bathingsoap'],label='bathingsoap')
plt.plot(data['month_number'],data['shampoo'],label='shampoo')
plt.plot(data['month_number'],data['moisturizer'],label='moisturizer')
plt.xlabel("Month")
plt.ylabel("Units Sold")
plt.legend()
```

```python
df2=data.drop(['total_profit','total_units'],axis=1)
df2.plot(kind='bar')

plt.xlabel('Months', fontweight ='bold', fontsize = 10,)
plt.ylabel('Sales', fontweight ='bold', fontsize = 10)
plt.legend()
plt.show()
```
