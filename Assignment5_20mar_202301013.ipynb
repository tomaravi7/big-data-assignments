```python
import pandas as pd
import seaborn as sb
import plotly.express as px
df=pd.read_csv('https://raw.githubusercontent.com/AnjulaMehto/MCA/main/company_sales_data.csv')
```

```python
print(df.head())
```

```python
sb.lineplot(df['total_profit'])
```

```python
product_sales = df.loc[:, 'facecream':'moisturizer']
sb.lineplot(data=product_sales)
```

```python
sb.barplot(data=df[df.columns.difference(['total_profit'])])
```

```python
fig = px.scatter(df)
fig.show()
```

```python
fig=px.bar(df,x='month_number',y='total_units')
fig.show()
```

```python
fig = px.histogram(df[df.columns.difference(['total_profit','total_units'])])
fig.show()
```
